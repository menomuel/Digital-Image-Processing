```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from deepface import DeepFace
```

```python
img_source = 'tests/yoda.jfif'
img = cv2.cvtColor(cv2.imread(img_source), cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(15, 15))
plt.subplot(1, 2, 1)
plt.axis('off')
plt.imshow(img)

try:
    img_face = DeepFace.detectFace(img_source, detector_backend="retinaface", align=False)
    plt.subplot(1, 2, 2)
    plt.axis('off')
    plt.imshow(img_face)
except:
    print("No face detected")
```

```python
analysis = DeepFace.analyze(img_source, enforce_detection=False, actions = ['age', 'gender', 'race', 'emotion'])

print("age:", analysis['age'])
print("gender:", analysis['gender'])
print("race:", max(analysis['race'].keys(), key=analysis['race'].get))
print("emotion:", max(analysis['emotion'].keys(), key=analysis['emotion'].get))

```
